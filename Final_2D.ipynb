```python
from Bio import SeqIO
from Bio import AlignIO
from Bio import Phylo

bat = SeqIO.read('RaTG13.fasta', 'fasta')
delta = SeqIO.read('delta.fasta', 'fasta')
omicron = SeqIO.read('omicron.fasta', 'fasta')
```

```python
print(bat.description)
print(delta.description)
print(omicron.description)

bat.id='Bat Covid'
```

```python
viruses = SeqIO.write([bat,delta,omicron,], "viruses.fasta", "fasta")
#https://www.ebi.ac.uk/Tools/msa/muscle/
#upload new file here
```

```python
with open("viruses.aln","r") as aln: 
    alignment = AlignIO.read(aln,"clustal")
print(type(alignment))
```

```python
# Open and initiate the Distance Calculator using the Identity model 
from Bio.Phylo.TreeConstruction import DistanceCalculator 
calculator = DistanceCalculator('identity')
```

```python
# Write the Distance Matrix 
distance_matrix = calculator.get_distance(alignment)
print(distance_matrix)
```

```python
# Open and initiate the Tree Constructor 
from Bio.Phylo.TreeConstruction import DistanceTreeConstructor
constructor = DistanceTreeConstructor(calculator)
```

```python
# Build the tree 
turtle_tree = constructor.build_tree(alignment)
turtle_tree.rooted = True
print(turtle_tree)
```

```python
# Save the tree to a new file 
Phylo.write(turtle_tree, "viruses_tree.xml", "phyloxml")
```

```python
# Import matplotlib and create a basic tree 
import matplotlib
import matplotlib.pyplot as plt
fig = Phylo.draw(turtle_tree)
```
